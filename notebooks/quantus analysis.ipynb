```python
import pandas as pd
```

```python
df = pd.read_csv('fastsverl_metrics.csv')
df
```

```python
df = df[~((df['metric_type'] == 'Faithfulness') & (df['explanation_type'] == 'value'))]
df
```

```python
metric_order = [
    'FaithfulnessCorrelation',
    'PixelFlipping',
    'Selectivity',
    'AvgSensitivity',
    'MaxSensitivity',
    'LocalLipschitzEstimate'
]
```

```python
avg_by_metric = df.groupby('metric_name')['value'].mean()
print(avg_by_metric.loc[metric_order])
```

```python
avg_by_metric_type = df.groupby('metric_type')['value'].mean()
print(avg_by_metric_type)
```

```python
avg_by_agent = df.groupby(['metric_name', 'agent'])['value'].mean().unstack()
print(avg_by_agent.loc[metric_order])
```

```python
avg_by_env = df.groupby(['metric_name', 'environment'])['value'].mean().unstack()
print(avg_by_env.loc[metric_order])
```

```python
avg_by_method = df.groupby(['metric_name', 'method_index'])['value'].mean().unstack()
print(avg_by_method.loc[metric_order])
```

```python
avg_by_method = df.groupby(['metric_name', 'explanation_type'])['value'].mean().unstack()
print(avg_by_method.loc[metric_order])
```
