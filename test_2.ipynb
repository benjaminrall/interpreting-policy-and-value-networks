```python
import torch
```

```python
from src import Trainer, Trainable
from src.sverl import Characteristic
```

```python
trainer = Trainer.load_checkpoint('checkpoints\\breakout-ppo-pc-2a-1744757310\\200.pt')
# trainer = Trainer.load_checkpoint('models\\breakout\\pc.pt')
```

```python
model: Characteristic = trainer.objective

samples, masks = model.generate_validation_data()
# Measures validation loss
total_loss = 0
for i, (x, y) in enumerate(samples):
    model.masker.mask(x, masks[i])
    predictions = model.model(x)
    if 1 in x:
        print("ALERT")
    if model.cfg.target == 'actor':
        y = torch.softmax(y, dim=-1)
        predictions = torch.softmax(predictions, dim=-1)
    print([round(e.item(), 4) for e in y[0]], [round(e.item(), 4) for e in predictions[0]])
    total_loss += torch.square(y - predictions).sum().item()

# Calculates and logs the average validation loss
total_loss /= model.cfg.validation_samples
print(total_loss)
```
